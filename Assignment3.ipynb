##### 7.  You have a product code that follows the format "SKU-12345-XYZ".

##### Find the first position of the dash (-).

##### Count the number of dashes (-) in the product code.

##### Remove the "SKU-" prefix and "-XYZ" suffix from the product code, Extract the numeric part 12345

##### Hint: Use String Methods

```python
code="SKU-12345-XYZ"
codefirst_position=code.find("-")
dash_count=code.count("-")
remove_prefix_suffix=code.replace("SKU","").replace("XYZ","")
print("first_position of dash:",codefirst_position)
print("Dash Count:",dash_count)
print("Remove prefix Suffix",remove_prefix_suffix)
```

##### 9. text = "    Data Analysis with Python is very powerful    "

##### Perform the following tasks:

##### Remove extra spaces using strip().
##### Convert the string into title case using title().
##### Replace "Powerful" with "Important".
##### Find the position of the word "Python".

```python
text = " Data Analysis with Python is very powerful "
remove_extra_space=text.strip()
str_to_title=text.title()
text_replace=text.replace("powerful","Important")
find_position=text.find("Python")
print(remove_extra_space)
print(str_to_title)
print(text_replace)
print("Python Position:",find_position)
```

##### 10. Given the list:
##### data = [12, 45, 2, 67, 34, 89, 23, 90]

##### Perform the following tasks:

##### Sort the list in ascending order.
##### Remove the highest value from the list.
##### Insert the value 50 at index 3.".
##### Finally, reverse the updated list and display the result.

```python
data = [12, 45, 2, 67, 34, 89, 23, 90]
data.sort()
print("Sorted List:",data)
data.pop()
print("After Removing Highest Value form The List:",data)
data.insert(3,50)
print("After Insertion:",data)
data.reverse()
print("Reversed the Updated List is:",data)
```
